```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("test 2.csv")  
print(df.shape)
df.head()
print(df.isnull().sum())
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Cabin"] = df["Cabin"].fillna("Unknown")
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
print(df.describe(include='all'))
print(df.info())
```

```python
# Visualize PassengerId
sns.countplot(x='PassengerId', data=df.head(10))
plt.title("Id of the passengers")
plt.show()
```

```python
# Passenger details:Name and gender
sns.countplot(x='Name', hue='Sex', data=df.head(8))
plt.title("Passenger Details: Name and Sex")
plt.xticks(rotation=45) 
plt.legend(loc='upper right')  
plt.show()
```

```python
#Age distribution
sns.histplot(df['Age'])
plt.title("Age Distribution")
plt.show()
```

```python

#Distribution of passenger class
pclass_counts = df['Pclass'].value_counts().sort_index()
labels = ['1st Class', '2nd Class', '3rd Class']
plt.figure(figsize=(6, 6))
plt.pie(pclass_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99','#ff9999'])

plt.title("Distribution of Passenger Classes ")
plt.show()
```
